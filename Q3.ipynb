```python
def update_dictionary(dct, key, value):
  if key in dct:
      print("Original value:", dct[key])

  dct[key] = value

  return dct

print(update_dictionary({}, "name", "Alice"))

print(update_dictionary({"age": 25}, "age", 26))
```
