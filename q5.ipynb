```python
def check_divisibility(num, divisor):
    if not (isinstance(num, (int, float)) and isinstance(divisor, (int, float))):
      return -1

    if divisor == 0:
      return False

    return num % divisor == 0

print(check_divisibility(10, 2))

print(check_divisibility(7, 3))
```
