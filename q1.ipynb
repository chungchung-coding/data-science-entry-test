```python
def swap(x, y):

    if not isinstance(x, (int, float)) or not isinstance(y, (int, float)):
        return -1

    x, y = y, x

    print("x =", x)
    print("y =", y)

print(swap("Apple", 10))

swap(9, 17)
```
